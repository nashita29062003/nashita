```python
import matplotlib.pyplot as plt
def create_pie_chart(labels, sizes):
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.show()
labels = ['Category 1', 'Category 2', 'Category 3 ', 'Category 4 ']
sizes = [30, 25, 20, 25]
create_pie_chart(labels, sizes)
```
