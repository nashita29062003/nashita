```python
import matplotlib.pyplot as plt
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 3, 5, 6]
plt.scatter(x_values ,y_values)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```
